```python
import cv2
from ultralytics import YOLO
# Load the YOLOv8 model
model = YOLO('models/ocean_echoes.pt')
```

```python
import cv2
from ultralytics import YOLO
from tqdm import tqdm


# Open the video file
video_path = "test_img/test.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Create VideoWriter object for output video
output_path = "output.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Adjust codec as needed
output_video = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Create tqdm progress bar
pbar = tqdm(total=total_frames, desc="Processing frames")


# Read until video is completed
while(cap.isOpened()):
    # Read a frame from the video
    success, frame = cap.read()
    if success == True:
        results = model(frame)
        # Visualize the results on the frame
        annotated_frame = results[0].plot()
        # Write the annotated frame to output video
        output_video.write(annotated_frame)
        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)
        # Update progress bar
        pbar.update(1)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

# When everything done, release the video capture object
cap.release()
cv2.destroyAllWindows()
output_video.release()

```

```python
import cv2
from ultralytics import YOLO
from tqdm import tqdm



# Open the video file
video_path = "test.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Create VideoWriter object for output video
output_path = "output.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Adjust codec as needed
output_video = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Create tqdm progress bar
pbar = tqdm(total=total_frames, desc="Processing frames")

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Write the annotated frame to output video
        output_video.write(annotated_frame)

        # Update progress bar
        pbar.update(1)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and release the output video
cap.release()
output_video.release()

# Close the tqdm progress bar
pbar.close()
```
